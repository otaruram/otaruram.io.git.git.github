```python


from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
import os

%matplotlib inline
```

```python
#Menyiapkan folder utama
BASIS_DIR = "/content/rockpaperscissors/rps-cv-images/"
#Image generator
training_datagen = ImageDataGenerator(
#Menormalisasi data
      rescale = 1./255,
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>AUGMENTASI<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
	    #Menerapkan skenario rotasi gambar hingga 40 derajat
      rotation_range=15,
      #Menerapkan skenario pelebaran gambar hingga hingga sebanyak 0.2 dari ukuran asli
      width_shift_range=0.2,
      #Menerapkan skenario peninggian gambar hingga hingg sebanyak 0.2 dari ukuran asli
      height_shift_range=0.2,
      #Menerapkan skenario pemangkasan gambar gambar hingga sebanyak 0.2 dari ukuran asli
      shear_range=0.2,
      zoom_range=0.2,
      #Menerapkan skenario putar balik gambar secara horizontal
      horizontal_flip=True,
      #Mengisi piksel yang kosong dengan nilai piksel terdekat
      fill_mode='reflect',
#>>>>>>>>>>>MEMECAH DATA MENJADI DATA TRAINING DAN DATA VALIDASI<<<<<<<<<<<<<<<<
     #80% Data Training, 20% Data Validasi
      validation_split=0.4
)
```

```python
#Generator data training
train_generator = training_datagen.flow_from_directory(
	#Folder target untuk digenerasi
  BASIS_DIR,
	#Mengubah resolusi seluruh gambar menjadi 150x150 piksel
 target_size=(150,150),
	#Kelas data kategori untuk klasifikasi multivariabel
 class_mode='categorical',
  shuffle=True,
  #Untuk data training
  subset='training'
  )

#Generator data validasi
validation_generator = training_datagen.flow_from_directory(
	#Folder target untuk digenerasi
  BASIS_DIR,
	#Mengubah resolusi seluruh gambar menjadi 150x150 piksel
 target_size=(150,150),
	#Kelas data kategori untuk klasifikasi multivariabel
 class_mode='categorical',
  shuffle=True,
  Untuk data validasi
 subset='validation'
  )
```

```python
#Membentuk model sequential
model = tf.keras.models.Sequential([
    #Bentuk input adalah gambar dengan ukuran 150x150 dengan 3 byte warna
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)), #Layer Konvolusi pertama
    #Layer Konvolusi kedua
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    #Layer Konvolusi ketiga
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
     #Layer input untuk Deep Learning Neural Network
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    #Hiden layer pertama dengan 512 neuron
    tf.keras.layers.Dense(128, activation='relu'),
    #Hiden layer kedua dengan 512 neuron
    tf.keras.layers.Dense(512, activation='relu'),
    #Layer output
    tf.keras.layers.Dense(3, activation='softmax')
])

model.summary()
```

```python
#Loss function Yang digunakan untuk Klasifikasi Kategori Lebih Dari 2
model.compile(loss = 'categorical_crossentropy',
#Fungsi optimizer(Yang Pernah diajarkan adalah 'adam')
              optimizer='rmsprop',
#Menampilkan akurasi model training
              metrics=['accuracy'])
```

```python
history=model.fit(
    #Data training hasil generasi
    train_generator,
    #Jumlah batch yang akan dieksekusi pada setiap epoch
    steps_per_epoch=4,
    #Jumlah perulangan komputasi(epoch)
    epochs=25,
    #Menampilkan akurasi pengujian data validasi
    validation_data=validation_generator,
    #Jumlah batch yang akan dieksekusi pada setiap epoch
    validation_steps=4,
    verbose=2
    )
```

```python
#Mengimpor modul matplotlib
import matplotlib.pyplot as plt
#Nilai akurasi training
acc = history.history['accuracy']
#Nilai akurasi validasi
val_acc = history.history['val_accuracy']
#Nilai loss training
loss = history.history['loss']
#Nilai loss validasi
val_loss = history.history['val_loss']
#Jumlah epoch
epochs = range(len(acc))
#Plot nilai akurasi training
plt.plot(epochs, acc, 'r', label='Akurasi Training ')
#Plot nilai akurasi validasi
plt.plot(epochs, val_acc, 'b', label='Akurasi Validasi')
#Menamai gambar
plt.title('Akurasi Training dan Validasi')
#Menampilkan legenda
plt.legend(loc=0)
plt.figure()
plt.show()
```

```python
#*************************KRITERIA HENTI TRAINING MODEL*************************
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('val_accuracy')>0.9):
      print("\nPELATIHAN BERHENTI, AKURASI MODEL SUDAH LEBIH DARI 90%!")
      self.model.stop_training = True

callbacks = myCallback()

#********************************MEMBANGUN MODEL********************************
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Dense(128, activation='relu'),

    tf.keras.layers.Dense(512, activation='relu'),

    tf.keras.layers.Dense(3, activation='softmax')
])

model.summary()

#******************************MENGKOMPILASI MODEL******************************
model.compile(loss = 'categorical_crossentropy',
optimizer='rmsprop', metrics=['accuracy'])

#*********************************MELATIH MODEL*********************************
history=model.fit(
    train_generator,
    steps_per_epoch=16,
    epochs=100,
    validation_data=validation_generator,
    validation_steps=4,
    verbose=2,
    callbacks=[callbacks]
    )

#************************PLOT AKURASI TRAINING & VALIDASI***********************
import matplotlib.pyplot as plt
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend(loc=0)
plt.figure()
plt.show()
```
